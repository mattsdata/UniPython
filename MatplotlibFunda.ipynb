## Matplotlib Practise

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)
```

```python
# Display the first few rows of the dataset
titanic.head()
```

```python
# Line Plot
plt.plot(titanic['Age'])
plt.title("Line Plot")
plt.show()
```

```python
# Scatter Plot
plt.scatter(titanic['Age'], titanic['Fare'], c=titanic['Survived'], cmap='coolwarm', alpha=0.6)
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.colorbar(label="Survived")
plt.show()
```

```python
# Bar Chart
titanic['Pclass'].value_counts().sort_index().plot(kind='bar')
plt.title("Bar Chart")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
# Histogram
plt.hist(titanic['Age'].dropna(), bins=20)
plt.title("Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie Chart
titanic['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart")
plt.ylabel("")
plt.show()
```

```python
# Box Plot
sns.boxplot(data=titanic, x='Pclass', y='Age')
plt.title("Box Plot")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Violin Plot
sns.violinplot(data=titanic, x='Pclass', y='Age')
plt.title("Violin Plot")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Heatmap
sns.heatmap(titanic.corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

```python
# Contour Plot
plt.contourf(titanic.pivot_table(index='Pclass', columns='Sex', values='Fare'))
plt.title("Contour Plot")
plt.xlabel("Sex")
plt.ylabel("Passenger Class")
plt.colorbar(label="Fare")
plt.show()
```

```python
# 3D Plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(titanic['Age'], titanic['Fare'], titanic['Pclass'], c=titanic['Survived'], cmap='coolwarm', alpha=0.6)
ax.set_title("3D Plot")
ax.set_xlabel("Age")
ax.set_ylabel("Fare")
ax.set_zlabel("Passenger Class")
plt.show()
```

```python
# Error Bars
sns.barplot(data=titanic, x='Pclass', y='Age', errorbar='sd')
plt.title("Error Bars")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

```python
# Polar Plot
angles = [0, 90, 180, 270]
values = titanic.groupby('Pclass')['Fare'].mean().tolist()
angles += angles[:1]
values += values[:1]
plt.subplot(111, polar=True)
plt.plot(angles, values)
plt.title("Polar Plot")
plt.show()
```

```python
# Stream Plot
plt.streamplot(range(len(titanic)), titanic['Age'], titanic['Fare'], color=titanic['Survived'], cmap='coolwarm')
plt.title("Stream Plot")
plt.xlabel("Index")
plt.ylabel("Age")
plt.colorbar(label="Survived")
plt.show()
```

```python
# Quiver Plot
plt.quiver(range(len(titanic)), titanic['Age'], titanic['Fare'], titanic['Pclass'])
plt.title("Quiver Plot")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```

```python
# Stem Plot
plt.stem(range(len(titanic)), titanic['Age'], linefmt='-', markerfmt='o', basefmt=' ')
plt.title("Stem Plot")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```

```python
# Area Plot
titanic.groupby('Pclass')['Age'].mean().plot(kind='area')
plt.title("Area Plot")
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")
plt.show()
```

```python
# Stack Plot
titanic.groupby('Pclass')['Age'].mean().plot(kind='bar', stacked=True)
plt.title("Stack Plot")
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")
plt.show()
```

```python
# Hexbin Plot
plt.hexbin(titanic['Age'], titanic['Fare'], gridsize=20)
plt.title("Hexbin Plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.colorbar(label="Frequency")
plt.show()
```

```python

# Waterfall Plot
values = [titanic['Age'].mean(), titanic['Fare'].mean(), titanic['Pclass'].mean()]
plt.bar(range(len(values)), values, color='b', alpha=0.5)
plt.title("Waterfall Plot")
plt.ylabel("Mean Value")
plt.grid(True)
plt.show()
```

```python

# Radial Bar Chart
angles = [0, 90, 180, 270]
values = titanic.groupby('Pclass')['Fare'].mean().tolist()
angles += angles[:1]
values += values[:1]
ax = plt.subplot(111, polar=True)
ax.fill(angles, values, 'b', alpha=0.1)
ax.plot(angles, values, 'b', linewidth=2)
plt.title("Radial Bar Chart")
plt.show()
```

```python
# Spider Plot (Radar Chart)
labels = ['Pclass 1', 'Pclass 2', 'Pclass 3']
stats = titanic.groupby('Pclass')['Fare'].mean().tolist()
stats = np.concatenate((stats,[stats[0]]))
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, stats, color='b', alpha=0.25)
ax.plot(angles, stats, color='b', linewidth=2)
plt.title("Spider Plot (Radar Chart)")
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
plt.show()
```
